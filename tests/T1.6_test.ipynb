```python
#PYTHONPATH=src python -c "
import pandas as pd
from cragb.data.load import sample_dataframe
from cragb.data.features import add_review_features
from cragb.utils.io import load_config

cfg = load_config('configs/data.yaml')
df = sample_dataframe(cfg['paths']['reviews_raw'], 20000)

with_img = df[df['images'].map(len) > 0].head(3)
without_img = df[df['images'].map(len) == 0].head(2)
sample = pd.concat([with_img, without_img])
featured = add_review_features(sample)

for i, row in featured.iterrows():
    raw_text = df.loc[i, 'text']
    raw_images = df.loc[i, 'images']
    print('row', i, '| token_len:', row['token_len'], '| naive word count:', len(raw_text.split()))
    print('  raw images:', len(raw_images), '-> has_image:', row['has_image'], 'n_images:', row['n_images'])
    if raw_images:
        assert raw_images[0].get('large_image_url') == row['image_urls'][0]
        print('  URL match OK:', row['image_urls'][0])
print('ALL CHECKS PASSED')
#"
```
